2. Generate a discrete uniform distribution and plot the PMF

A discrete uniform distribution means all values in a given range have equal probability. We can use numpy and matplotlib to generate and plot its Probability Mass Function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the range
low, high = 1, 6  # Simulating a fair die roll
values = np.arange(low, high + 1)

# Compute PMF
pmf = stats.randint.pmf(values, low, high + 1)

# Plot PMF
plt.bar(values, pmf, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.show()
```

3. Write a Python function to calculate the Probability Density Function (PDF) of a Bernoulli distribution

A Bernoulli distribution models a binary outcome (success/failure, 1/0) with probability . The Probability Density Function (PDF) is given by:

P(X = x) = p^x (1 - p)^{(1-x)}

where  is either 0 or 1.

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example usage
p = 0.5  # Probability of success
x_values = [0, 1]

# Compute PDF values
pdf_values = [bernoulli_pdf(p, x) for x in x_values]

print("Bernoulli PDF values:", pdf_values)
```

4. Simulate a Binomial Distribution and Plot Its Histogram

A binomial distribution represents the number of successes in  independent trials, where each trial has success probability .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of simulations

# Generate binomial data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2)-0.5, density=True, alpha=0.6, color='b', edgecolor='black')

# Overlay PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'ro', markersize=8, label="PMF")

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.legend()
plt.show()
```

6. Calculate and Plot the Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution

The Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a given value.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for a fair die (1 to 6)
low, high = 1, 6
x = np.arange(low, high + 2)  # Extra value for stepwise plotting

# Compute CDF
cdf = randint.cdf(x, low, high + 1)

# Plot
plt.step(x, cdf, where='post', color='blue', label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution')
plt.legend()
plt.grid()
plt.show()
```

8. Simulate Data from a Normal Distribution and Plot Its Histogram

A normal distribution (Gaussian) is defined by a mean  and standard deviation .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu, sigma = 0, 1  # Mean and standard deviation
size = 1000  # Number of samples

# Generate normal distribution data
data = np.random.normal(mu, sigma, size)

# Plot histogram
plt.hist(data, bins=30, density=True, color='purple', edgecolor='black', alpha=0.6)

# Overlay theoretical PDF
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r', label='PDF')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```

9. Calculate Z-Scores from a Dataset and Plot Them

A Z-score measures how many standard deviations a data point is from the mean:

Z = \frac{X - \mu}{\sigma}

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
data = np.random.normal(50, 10, 1000)  # Mean=50, Std Dev=10

# Calculate Z-scores
z_scores = (data - np.mean(data)) / np.std(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, density=True, color='brown', edgecolor='black', alpha=0.6)
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Z-Score Distribution')
plt.show()
```

10. Implement the Central Limit Theorem (CLT) Using Python

The Central Limit Theorem states that the sampling distribution of the mean approaches normality as the sample size increases, regardless of the population's distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 30  # Each sample contains 30 values
num_samples = 1000  # Number of samples
population = np.random.exponential(scale=2, size=10000)  # Non-normal population

# Compute sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, color='blue', edgecolor='black', alpha=0.6)
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem Demonstration')
plt.show()
```

19. Implement Hypothesis Testing Using Z-Statistics

This checks if a sample mean significantly differs from the population mean.

```python
from scipy.stats import norm

# Sample mean, population mean, population std deviation, sample size
sample_mean = 52
population_mean = 50
population_std = 10
sample_size = 30

# Calculate Z-score
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f'Z-score: {z}')
print(f'P-value: {p_value}')
```

20. Create a Confidence Interval for a Dataset

A confidence interval estimates the range in which the true mean lies.

```python
import scipy.stats as stats

# Generate sample data
data = np.random.normal(50, 10, 100)

# Compute confidence interval (95%)
confidence = 0.95
mean = np.mean(data)
sem = stats.sem(data)
margin = sem * stats.t.ppf((1 + confidence) / 2, len(data) - 1)

ci_lower = mean - margin
ci_upper = mean + margin

print(f'95% Confidence Interval: ({ci_lower}, {ci_upper})')
```
